```python
import nibabel as nb
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Mirar una imagen de CT simple
img_path = '/data/Datasets/stroke/FOSCAL/ACV/ACV-005/TC_20211122/nifti.gz/Head 7.0 Axial.1/SER00002_CEREBRAL_5&10mm_20211122154232_11.nii.gz'

nib_img = nb.load(img_path)
slices = nib_img.get_fdata()
print(slices.shape)

# Se plotean los slices
plt.figure(figsize=(20, 20))

for idx in range(slices.shape[-1]):
    plt.subplot(5, 5, idx + 1)
    plt.imshow(slices[:, :, idx])
plt.show()
```
