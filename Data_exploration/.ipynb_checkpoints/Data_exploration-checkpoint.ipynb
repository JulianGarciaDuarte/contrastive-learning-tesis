# Images analysis

```python
def set_patch_text(ax, ax_patch_idx, text, offset, is_sub_text=False):
    patch = ax.patches[ax_patch_idx]
    if is_sub_text:
        ax.text(patch.get_x() + offset[0],
                patch.get_height()/2 + offset[1],
                text,
                color='white',
                size='small')
    else:
        ax.text(patch.get_x() + offset[0],
                patch.get_height()/2 + offset[1],
                text)

def remove_outliers(data, outlier_constant=1.5):
    data = np.array(data)
    upper_q = np.percentile(data, 75)
    lower_q = np.percentile(data, 25)
    result = []
    
    # The interquatile range
    IQR = (upper_q - lower_q) * outlier_constant
    quartile_set = [lower_q - IQR, upper_q + IQR]
    
    for sample in data.tolist():
        if sample >= quartile_set[0] and sample <= quartile_set[1]:
            result.append(sample)
            
    return np.array(result)

def normalize_text(text):
    """
    Removes all accents and returns the given text in lower case.
    """
    base='áéíóú'
    result='aeiou'
    mapping_table = base.maketrans(base, result)
    return text.translate(mapping_table).lower().strip()
```

# Plan for the data analisis

1. Plot per sex, age

```python
acv_data = pd.read_csv('Base de datos pacientes Stroke - ACV.csv', header=1)
acv_data.columns = ['Paciente', 'Fecha de ingreso', 'TC simple', 'Angio TC', 'Resonancia',
       'Otros estudios', 'Edad', 'Sexo', 'Hipertensión', 'Tabaquismo',
       'Diabetes', 'Dislipidemis', 'Fibrilación Auricular', 'ACV previo',
       'Otros antecedentes', 'Tiempo desde inicio de sintomas hasta ingreso ',
       'Hemiparesia', 'Hemiplejía', 'Afasia motora',
       'Desviación comisura labial', 'Alt del estado de conciencia',
       'Disartria', 'Cefalea', 'Otros', 'NIHSS', 'Tratamiento realizado',
       'Fecha tratamiento', 'Hallasgos CT sin contraste', 'ASPECTS TC', 'ASPECTS difusión',
       'Arteria comprometida', 'CBV', 'CBF absoluto', 'CBF relativo', 'Tmax',
       'TTP', 'T. Transito medio', 'Valor escala TICI', 'mRS egreso',
       'mRS 30 días', 'Notas']
acv_data['Edad'] = acv_data['Edad'].replace('N', np.nan).astype('float64', errors='ignore')
```

```python
identificadores = ['Paciente',
                         'Edad',
                         'Sexo'
                  ]
columnas_demograficas = ['Hipertensión', 'Tabaquismo', 'Diabetes', 'Dislipidemis', 'Fibrilación Auricular', 'ACV previo']
datos_demograficos = acv_data[[col for col in columnas_demograficas + identificadores]]
demo_data = datos_demograficos.dropna(how='any', subset=columnas_demograficas).reset_index()
```

```python
antecedentes = ['Hipertensión', 'Tabaquismo', 'Diabetes', 'Dislipidemis', 'Fibrilación Auricular', 'ACV previo']
cuenta_lista = {ant: [] for ant in antecedentes}
CODIGO_POSITIVO = 1.0
CODIGO_NEGATIVO = 2.0

for ant in antecedentes:
    positivos = (acv_data[ant] == CODIGO_POSITIVO).sum()
    negativos = (acv_data[ant] == CODIGO_NEGATIVO).sum()
    nan = acv_data[ant].size - (positivos + negativos)
    
    cuenta_lista[ant].append(positivos)
    cuenta_lista[ant].append(negativos)
    cuenta_lista[ant].append(nan)
    cuenta_lista[ant].append(nan + positivos + negativos)


demo_resum = pd.DataFrame(data=cuenta_lista)
demo_resum.index = ['Positivos', 'Negativos', 'Sin Datos', 'Total']
demo_resum
```

```python
# Encuentra los otros antecedentes mas comunes y su frecuencia
otros_ant = {}
for ant in acv_data['Otros antecedentes']:
    ant = str(ant)
    if ant == 'nan':
        continue
        
    splitted = ant.split(',')
    for mot in splitted:
        norm = normalize_text(mot)
        if norm not in otros_ant.keys():
            otros_ant[norm] = 1
        else:
            otros_ant[norm] += 1

    
resum_otros_ant = pd.DataFrame(data=[otros_ant])
resum_otros_ant.index=['Frecuencia']
resum_otros_ant
```

# Plots

## 1. Discriminando por sexo

```python
SEX_CODE_MALE = 1.0
SEX_CODE_FEMALE = 2.0
```

```python
sex_info = acv_data['Sexo']
n_nan = np.isnan(sex_info).sum()
n_male = (sex_info == SEX_CODE_MALE).sum()
n_female = (sex_info == SEX_CODE_FEMALE).sum()
total = n_nan + n_male + n_female
per_nan = n_nan/total
per_sex = (n_male + n_female)/total
per_male = n_male/(total)
per_female = n_female/(total)

#fig, ax = plt.subplots(figsize=(6, 7))
plt.figure(figsize=(12, 7))

# El grafico de barras
ax = plt.subplot(1, 2, 1)
# Agrega los plots
ax.bar(['Sin informacion'], [n_nan], color='darkgray')
ax.bar(['Sexo'], [n_male], color='steelblue')
ax.bar(['Sexo'], [n_female], bottom=[n_male], color='darkorange')

# Agrega las anotaciones de procentajes globales
# Nan
offset = [0.31, 65]
set_patch_text(ax, 0, f'{per_nan * 100:.2f}%', offset)

# Sex
offset = [0.31, 40]
set_patch_text(ax, 1, f'{per_sex * 100:.2f}%', offset)

# Male 
offset = [0.31, -2]
set_patch_text(ax, 1, f'{per_male * 100:.2f}%', offset, is_sub_text=True)

# Female
offset = [0.31, 23]
set_patch_text(ax, 1, f'{per_female * 100:.2f}%', offset, is_sub_text=True)

ax.set_ylim([0, 140])
ax.legend([f'Sin informacion ({n_nan})', f'Masculino ({n_male})', f'Femenino ({n_female})'])
ax.set_title('Pacientes discriminados por sexo')

# El pie chart
ax = plt.subplot(1, 2, 2)
sizes = [n_male, n_female]
labels = ['Masculino', 'Femenino']
ax.pie(sizes, labels=labels, autopct='%1.2f%%')
ax.set_title('Distribución de pacientes segun su sexo')


plt.show()
```

## 2. Discriminando por sexo y edad

```python
edad_info = acv_data[np.isnan(acv_data['Edad']) == 0]
age_male = edad_info[edad_info['Sexo'] == SEX_CODE_MALE].loc[:, ('Edad')].values
age_female = edad_info[edad_info['Sexo'] == SEX_CODE_FEMALE].loc[:, ('Edad')].values
n_bins_fem = 10
n_bins_male =int((age_male.size*n_bins_fem)/age_female.size)

# Se crean la figura y el layout
fig = plt.figure(figsize=(10, 9))
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Se crean los ejes
boxplot_ax = fig.add_subplot(grid[:-1, :-1])
hist_ax = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])

# Se plotea el boxplot
boxplot_ax.boxplot([age_male, age_female])
boxplot_ax.set_xticklabels(['Masculino', 'Femenino'])

# Se plotea el histograma
hist_ax.hist(age_female,
             orientation='horizontal',
             bins=n_bins_fem,
             fc=(0, 0, 0, 1))
hist_ax.hist(age_male,
             orientation='horizontal',
             bins=n_bins_male,
             fc=(1, 0.5, 0, 0.6))

for idx, patch in enumerate(hist_ax.patches):
    patch.set_edgecolor((0, 0, 0))
    
hist_ax.legend(['Femenino', 'Masculino'])

fig.suptitle('Distribucion de edad en pacientes discriminados por sexo')
plt.show()
```

## 2.1 Disciminando por sexo y edad quitando outliers

```python
edad_info = acv_data[np.isnan(acv_data['Edad']) == 0]
age_male = edad_info[edad_info['Sexo'] == SEX_CODE_MALE].loc[:, ('Edad')].values
age_female = edad_info[edad_info['Sexo'] == SEX_CODE_FEMALE].loc[:, ('Edad')].values
age_male = remove_outliers(age_male)
age_female = remove_outliers(age_female)

n_bins_fem = 10
n_bins_male =int((age_male.size*n_bins_fem)/age_female.size)

# Se crean la figura y el layout
fig = plt.figure(figsize=(10, 9))
grid = plt.GridSpec(4, 4, hspace=0.5, wspace=0.2)

# Se crean los ejes
boxplot_ax = fig.add_subplot(grid[:-1, :-1])
hist_ax = fig.add_subplot(grid[:-1, -1], xticklabels=[], yticklabels=[])

# Se plotea el boxplot
boxplot_ax.boxplot([age_male, age_female])
boxplot_ax.set_xticklabels(['Masculino', 'Femenino'])

# Se plotea el histograma
hist_ax.hist(age_female,
             orientation='horizontal',
             bins=n_bins_fem,
             fc=(0, 0, 0, 1))
hist_ax.hist(age_male,
             orientation='horizontal',
             bins=n_bins_male,
             fc=(1, 0.5, 0, 0.6))

for idx, patch in enumerate(hist_ax.patches):
    patch.set_edgecolor((0, 0, 0))
    
hist_ax.legend(['Femenino', 'Masculino'])

fig.suptitle('Distribucion de edad en pacientes discriminados por sexo (sin outliers)')
plt.show()
```
